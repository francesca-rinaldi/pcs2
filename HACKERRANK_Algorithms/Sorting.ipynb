```python
#Correctness and the Loop Invariant
def insertion_sort(l):
    for i in range(1, len(l)):
        j = i-1
        key = l[i]
        while (j >= 0) and (l[j] > key):
           l[j+1] = l[j]
           j -= 1
        l[j+1] = key

m = int(input().strip())
ar = [int(i) for i in input().strip().split()]
insertion_sort(ar)
print(" ".join(map(str,ar)))
```

```python
#Quicksort 1 - Partition
import math
import os
import random
import re
import sys

def quickSort(arr):
    p = arr[0]
    left = []
    right = []
    for e in arr[1:]:
        if e <= p:
            left.append(e)
        else:
            right.append(e)
    left.append(p)
    left.extend(right)
    return(left)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    n = int(input())
    arr = list(map(int, input().rstrip().split()))
    result = quickSort(arr)
    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')
    fptr.close()
```

```python
#Counting Sort 1
import os
import random
import re
import sys

def countingSort(arr):
    lst = [0] * 100
    for i in range(len(lst)):
        if i in arr:
            lst[i] = arr.count(i)
    return(lst)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    n = int(input())
    arr = list(map(int, input().rstrip().split()))
    result = countingSort(arr)
    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')
    fptr.close()
```

```python
#Counting Sort 2
import math
import os
import random
import re
import sys

def countingSort(arr):
    lst = [0] * 100
    val = []
    for i in range(len(lst)):
        if i in arr:
            lst[i] = arr.count(i)
    for j in range(len(lst)):
        if lst[j] > 0:
            val.extend([j] * lst[j])
    return(val)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    n = int(input())
    arr = list(map(int, input().rstrip().split()))
    result = countingSort(arr)
    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')
    fptr.close()
```
