```python
#Diagonal Difference
import math
import os
import random
import re
import sys

# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY arr as parameter.
#

def diagonalDifference(arr):
    d1 = []
    d2 = []
    for n in range(len(arr[0])):
        d1.append(int(arr[n][n]))
        d2.append(int(arr[n][-(n+1)]))
    return(abs(sum(d1)-sum(d2)))
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#Plus minus
import math
import os
import random
import re
import sys

def plusMinus(arr):
    positive = 0
    negative = 0
    zeros = 0
    for e in arr:
        if e > 0:
            positive += 1
        elif e < 0:
            negative += 1
        elif e == 0:
            zeros += 1
    print(float(positive/len(arr)))
    print(float(negative/len(arr)))
    print(float(zeros/len(arr)))

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().rstrip().split()))
    plusMinus(arr)
```

```python
#Staircase
import math
import os
import random
import re
import sys

# Complete the staircase function below.
def staircase(n):
    for i in range(1, n + 1):
        print(('#'*i).rjust(n))

if __name__ == '__main__':
    n = int(input())
    staircase(n)
```

```python
#Time conversion
import os
import sys

def timeConversion(s):
    hh = s[:2]
    mm = s[3:5]
    ss = s[6:8]
    ap = s[8:]
    if 'PM' in ap and int(hh) < 12:
        return(str(int(hh) + 12) + ':' + mm + ':' + ss)
    elif 'AM' in ap and int(hh) >= 12:
        return('00:' + mm + ':' + ss)
    else:
        return(s[:8])

if __name__ == '__main__':
    f = open(os.environ['OUTPUT_PATH'], 'w')
    s = input()
    result = timeConversion(s)
    f.write(result + '\n')
    f.close()
```
